```python
import pandas as pd
df1 = pd.read_csv("orders_features.csv")
df2 = pd.read_csv("orders_features.csv")
if "order_id" in df1.columns and "order_id" in df2.columns:
    combined_df = pd.merge(df1, df2, on="order_id", how="outer", suffixes=("", "_dup"))
    dup_cols = [col for col in combined_df.columns if col.endswith("_dup")]
    combined_df = combined_df.drop(columns=dup_cols)
else:
    shared_cols = [col for col in df2.columns if col in df1.columns]
    df2_filtered = df2.drop(columns=shared_cols)
    combined_df = pd.concat([df1, df2_filtered], axis=1)
combined_df.to_csv("cfeatures.csv", index=False)
print("Shape:", combined_df.shape)
print("Columns:", combined_df.columns.tolist())
```
