```python
import numpy as np, pandas as pd
rng = np.random.default_rng(42)
n_users, n_orders = 5000, 40000
users = pd.DataFrame({
"user_id": range(n_users),
"signup_date": pd.Timestamp("2023-01-01")+pd.to_timedelta(rng.integers(0, 1000, n_users), unit="D"),
"city": rng.choice(["Lahore","Karachi","Islamabad","Multan","Quetta"], n_users, p=[.3,.3,.2,.12,.08]),
"age": rng.normal(33, 9, n_users).clip(18, 70).round(),})
o = pd.DataFrame({
"order_id": range(n_orders),
"user_id": rng.integers(0, n_users, n_orders),
"weight_kg": rng.gamma(2, 2, n_orders).round(2),
"distance_km": rng.gamma(3, 60, n_orders).round(0),
"express_flag": rng.random(n_orders) < 0.25,
"quantity": rng.integers(1, 6, n_orders),})
o["price"] = (o.quantity * rng.gamma(3, 800, n_orders)).round(0)
o["shipping_cost"] = (40 + 2.1*o.weight_kg + 0.55*o.distance_km+0.004*o.weight_kg*o.distance_km + 90*o.express_flag+rng.normal(0, 25, n_orders)).round(0)        
# true pattern + noise
# churn label: driven by recency & velocity (so your models CAN find it)
users.to_csv("users.csv", index=False); o.to_csv("orders.csv", index=False)
```
